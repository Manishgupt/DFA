```python
"The above program is NFA to DFA conversion"

#points to remember before executing the code
'''
1)enter $ ,if any state for specific path is not reachable(means  $ represent the phie value)
2)enter comma separated value if multiple state is reachable for specific path (ex:- q0,q1 )
3)enter integer value for number of state(ex:- 2,3,4 this type of value)
4)enter integer value for number of path also 
'''

import pandas as pd
#input section
###########################################################

dic={}
no_of_state=int(input("enter the number of state:"))
no_of_path=int(input("enter the number of path:"))
print("_"*45)
for i in range(no_of_state):
    a=input(f"enter {i} state:")
    dic[a]=[]
    for j in range(no_of_path):
        b=input(f"enter path {j} for {a} state:")
        dic[a].append(b.split(","))
    print("_"*45)
initial_state=list(dic.keys())[0]
final_state_nfa=input("\n---->enter final state of nfa:")
print("_"*45)

############################################################
#copy initial state from nfa to dfa
dic1={list(dic.keys())[0]:list(dic.values())[0]}
################################################################
def func():
    i=0
    while(True):
        for j in range(no_of_path):
            try: 
                if([list(dic1.keys())[i]]!=list(dic1.values())[i][j]):
                    key=str(list(dic1.values())[i][j])
                    if((str([list(dic1.keys())[0]])!=key)) and (key not in list(dic1.keys())):
                        dic1[key]=[]
                        for l in range(no_of_path):
                            temp=[]
                            for k in list(dic1.values())[i][j]:
                                try: dic[k][l]
                                except:
                                    for i in range(no_of_path):
                                        dic1[key].append(key.replace("'",""))
                                        
                                temp.extend(dic[k][l])
                            value=list(set(temp))
                            if(("$" in value) and (['$']!=value)):
                                value.remove("$")
                            dic1[key].append(value)
                            temp.clear()
            except: return
        i+=1
func()

###################################################
final_state_dfa=[i for i in list(dic1.keys()) if final_state_nfa in i]
##################################################
print("\n\nThe Given NFA---->\n")
print(pd.DataFrame(dic).T)
print("_"*45)
print("\n\nNFA TO DFA---->\n")
print(pd.DataFrame(dic1).T)
print("_"*45)
print("\ninitial_state-->",initial_state)
print("_"*45)
print("\nfinal state of NFA-->",final_state_nfa)
print("_"*45)
print("\nfinal state of DFa-->",set(final_state_dfa))
```
